## Jupyter Notebooks

A Jupyter notebook combines code, output, and narrative text in a single document. This makes it the standard tool for reproducible data analysis — you can hand a notebook to a colleague and they can re-run your entire analysis from scratch.

The key idea is **cell-by-cell execution**: you run cells in sequence, and the notebook remembers the results. Code and output live in the same place — no copying results into a separate document.

### Code Cells in Action

A single code cell holds your logic **and** its output — no copy/paste into a slide deck.

```python
# Quarterly revenue analysis — code and result live together in one cell
revenue = {"Q1": 125_000, "Q2": 98_000, "Q3": 142_000, "Q4": 110_000}

annual = sum(revenue.values())
best_quarter = max(revenue, key=revenue.get)

print(f"Annual revenue: ${annual:,.0f}")
print(f"Quarterly avg:  ${annual / 4:,.0f}")
print(f"Best quarter:   {best_quarter} (${revenue[best_quarter]:,.0f})")
```

### Notebooks Tell a Story

The most effective notebooks follow a **3-cell pattern**: a markdown cell that explains the *why*, a code cell that does the work, and another markdown cell that interprets the result. This turns raw code output into a coherent analytical narrative.

Here's the pattern in action — the three cells below work together as a complete analysis section:

**Business Question:** Which subscription plan is most popular?

We'll look at the distribution of customers across plan tiers to identify where to focus retention efforts.

```python
subscriptions = {"Basic": 820, "Premium": 310, "Enterprise": 120}
total = sum(subscriptions.values())

for plan, count in subscriptions.items():
    bar = "█" * (count // 40)
    print(f"{plan:12s}: {count:4,} customers ({count/total:.0%})  {bar}")
```

**Finding:** Basic plan dominates at 66% of customers. Premium (25%) is the primary upsell opportunity — targeted Basic → Premium campaigns could meaningfully improve revenue.

## Lists

A **list** is an ordered, mutable sequence — use it when the order of items matters and the data will change over time. In a retail analytics context, daily transaction counts belong in a list: new days keep getting added, and you may need to correct data entry errors.

**Create:** use square brackets `[]`  
**Access:** use index position (starts at 0)  
**Modify:** append, remove, or reassign by index

```python
# Create a list of daily sales figures
daily_sales = [1842, 2105, 1956, 2341, 890]
print(daily_sales)
```

```python
# Get an item by position (index starts at 0)
daily_sales = [1842, 2105, 1956, 2341, 890]
print(daily_sales[1])   # second item
```

```python
# Append a new item to the end
daily_sales = [1842, 2105, 1956, 2341, 890]
daily_sales.append(3102)
print(daily_sales)
```

```python
# Change an existing item
daily_sales = [1842, 2105, 1956, 2341, 890]
daily_sales[1] = 2200   # correct a data entry error
print(daily_sales)
```

## Tuples

A **tuple** is an ordered, immutable sequence — use it when the data is fixed and should be protected from accidental changes. Store coordinates, RGB colors, database connection settings, or any record where the values should never change.

**Create:** use parentheses `()`  
**Access:** use index position (same as lists)  
**Modify:** you can't — that's the point

```python
# Create a tuple of a store's GPS coordinates
store_location = (39.1031, -84.5120)
print(store_location)
```

```python
# Get items by position — or unpack into named variables
store_location = (39.1031, -84.5120)
lat, lon = store_location
print(lat)
print(lon)
```

```python
# Try to change a value — tuples won't allow it
store_location = (39.1031, -84.5120)
store_location[0] = 40.0   # this will raise a TypeError
```

## Sets

A **set** is an unordered collection of unique values — it automatically discards duplicates. Use it for deduplication and fast membership testing. Sets don't support indexing (there's no guaranteed order), but `in` checks are nearly instant even for millions of items.

**Create:** use curly braces `{}` with values (no key-value pairs)  
**Access:** use `in` for membership testing — no index  
**Modify:** use `.add()` to add items

```python
# Create a set — duplicates are silently removed
subscribers = {"john@email.com", "sarah@email.com", "mike@email.com", "john@email.com"}
print(subscribers)
print(f"Count: {len(subscribers)}")   # 3, not 4
```

```python
# Check if an item exists (very fast even for huge sets)
subscribers = {"john@email.com", "sarah@email.com", "mike@email.com"}
print("sarah@email.com" in subscribers)
print("alex@email.com" in subscribers)
```

```python
# Add a new item
subscribers = {"john@email.com", "sarah@email.com", "mike@email.com"}
subscribers.add("alex@email.com")
print(subscribers)
```

## Dictionaries

A **dictionary** stores key-value pairs — it's Python's built-in lookup table. Use it when you need to retrieve values by name or ID rather than by position. Product catalogs, customer profiles, configuration settings, and any "look up X to get Y" problem are natural fits.

**Create:** use curly braces `{}` with `key: value` pairs  
**Access:** use `dict[key]` to retrieve a value  
**Modify:** assign to a key to add or update

```python
# Create a product price dictionary
prices = {"organic milk": 4.99, "whole grain bread": 3.29}
print(prices)
```

```python
# Get a value by key
prices = {"organic milk": 4.99, "whole grain bread": 3.29}
print(prices["organic milk"])
```

```python
# Add a new key-value pair
prices = {"organic milk": 4.99, "whole grain bread": 3.29}
prices["eggs"] = 2.89
print(prices)
```

```python
# Update an existing value
prices = {"organic milk": 4.99, "whole grain bread": 3.29}
prices["organic milk"] = 5.49
print(prices["organic milk"])
```

## Packages and Libraries

Python's real power comes from its ecosystem. The **standard library** ships with Python and is always available (`math`, `datetime`, `random`). **Third-party libraries** must be installed separately but dramatically extend what you can do.

The key insight is that production-quality libraries are written by experts, heavily tested, and optimized. Using them isn't laziness — it's engineering judgment.

### Build from Scratch vs. Use a Library

```python
# Manual correlation — a lot of code for one number
x = [2, 4, 7, 8, 10, 11, 14, 13]
y = [1, 3, 5, 7, 10, 9, 13, 13]

mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)
numerator = sum((a - mean_x) * (b - mean_y) for a, b in zip(x, y))
denominator = (sum((a - mean_x)**2 for a in x) * sum((b - mean_y)**2 for b in y)) ** 0.5
print(f"Manual:  {numerator / denominator:.6f}")

# numpy does it in one line
print(f"numpy:   {np.corrcoef(x, y)[0, 1]:.6f}")
```

### Standard Library

Already installed with Python — no `pip install` needed.

```python
# random: sampling and weighted selection
outreach = random.choices(["Email", "Call", "Text"], weights=[50, 30, 20], k=5)
print(outreach)
```

```python
# datetime: date arithmetic
today = datetime.date.today()
trial_end = today + datetime.timedelta(days=30)
print(f"Free trial ends: {trial_end}")
```

### Third-Party Libraries

Must be installed (e.g., `pip install numpy`), but dramatically extend Python's capabilities.

```python
# numpy: fast numerical computing across entire arrays
prices = [12.99, 8.50, 24.99, 5.75, 18.00]
print(f"Avg: ${np.mean(prices):.2f}  |  Std dev: ${np.std(prices):.2f}  |  Max: ${np.max(prices):.2f}")
```
