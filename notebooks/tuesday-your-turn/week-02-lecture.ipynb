## Jupyter Notebooks

A Jupyter notebook combines code, output, and narrative text in a single document. This makes it the standard tool for reproducible data analysis — you can hand a notebook to a colleague and they can re-run your entire analysis from scratch. The key idea is **cell-by-cell execution**: you run cells in sequence, and the notebook remembers the results.

```python
# Notebook cells run in the order you execute them, not top-to-bottom automatically.
# This is the most common source of confusion for new users.

# Imagine these three cells executed out of order:
revenue_q1 = 125_000
revenue_q2 = 98_000
revenue_q3 = 142_000
revenue_q4 = 110_000

# Annual total and average
annual_revenue = revenue_q1 + revenue_q2 + revenue_q3 + revenue_q4
quarterly_avg = annual_revenue / 4

print(f"Annual revenue:  ${annual_revenue:,.0f}")
print(f"Quarterly avg:   ${quarterly_avg:,.0f}")
print(f"Best quarter:    Q3 (${revenue_q3:,.0f})")

# Key rule: always use Kernel → Restart & Run All before sharing a notebook.
# This ensures your results are reproducible from top to bottom.
```

## Lists and Tuples

When you need to store a collection of values, Python gives you ordered sequences. **Lists** are mutable — you can add, remove, and change items. **Tuples** are immutable — once created, they cannot be modified. Choosing between them is a design decision: use a list when data will change over time, a tuple when data is fixed and should be protected.

In a retail analytics context: daily sales figures belong in a list (new days keep getting added), but a store's GPS coordinates belong in a tuple (they never change).

```python
# List: daily transaction counts for a store — grows as days pass
daily_transactions = [412, 389, 501, 478, 623, 715, 698]

# Lists support indexing, slicing, and mutation
print(f"First day:       {daily_transactions[0]} transactions")
print(f"Last day:        {daily_transactions[-1]} transactions")
print(f"Best 3 days:     {sorted(daily_transactions, reverse=True)[:3]}")

# Add this week's Sunday count
daily_transactions.append(544)
print(f"After adding Sunday: {len(daily_transactions)} days tracked")
print()

# Tuple: store location — fixed, never changes
store_location = (39.7641, -84.1880)   # Cincinnati coordinates
print(f"Store location:  lat={store_location[0]}, lon={store_location[1]}")

# Tuples unpack cleanly — a common pattern
latitude, longitude = store_location
print(f"Latitude:        {latitude}")
print(f"Longitude:       {longitude}")

# Attempting store_location[0] = 40.0 would raise a TypeError — that's the point
```

## Sets and Dictionaries

**Sets** store unordered collections of *unique* values — they automatically discard duplicates, making them ideal for deduplication and membership testing. **Dictionaries** store *key-value pairs* — they're Python's built-in lookup table, letting you retrieve values by name rather than position.

In business analytics: a set of unique customer emails ensures you don't send duplicate marketing messages; a dictionary of product prices lets you look up any price instantly without scanning a list.

```python
# Set: unique customers who visited this week — duplicates auto-removed
monday_visitors = {"C001", "C042", "C107", "C001", "C083"}   # C001 visited twice
tuesday_visitors = {"C042", "C055", "C201", "C107"}

print(f"Unique Monday visitors:   {len(monday_visitors)}")    # 4, not 5
print(f"Visited both days:        {monday_visitors & tuesday_visitors}")
print(f"Monday only:              {monday_visitors - tuesday_visitors}")
print(f"All unique visitors:      {len(monday_visitors | tuesday_visitors)}")
print()

# Dictionary: product catalog — fast lookup by product ID
product_prices = {
    "PROD001": {"name": "Organic Apples", "price": 2.99, "unit": "lb"},
    "PROD002": {"name": "Whole Milk",     "price": 3.49, "unit": "gal"},
    "PROD003": {"name": "Sourdough Bread","price": 4.29, "unit": "loaf"},
}

# Look up a specific product
product = product_prices["PROD002"]
print(f"Product lookup: {product['name']} — ${product['price']:.2f} per {product['unit']}")

# Add a new product
product_prices["PROD004"] = {"name": "Free-Range Eggs", "price": 5.99, "unit": "doz"}

# Summarize catalog
print(f"\nCatalog: {len(product_prices)} products")
for pid, info in product_prices.items():
    print(f"  {pid}: {info['name']:20s}  ${info['price']:.2f}")
```

## Packages and Libraries

Python's real power comes from its ecosystem. The **standard library** ships with Python and is always available (`math`, `datetime`, `random`). **Third-party libraries** must be installed separately but dramatically extend what you can do — `numpy` alone would take weeks to replicate from scratch.

The key insight is that production-quality libraries are written by experts, heavily tested, and optimized. Using them isn't laziness — it's engineering judgment. A data analyst who writes their own correlation function from scratch when `numpy` exists is like a chef who mills their own flour.

```python
# Standard library — always available, no installation needed

# math: precise mathematical operations
monthly_revenue = 847_293.45
growth_rate = 0.073
months = 12
projected = monthly_revenue * math.pow(1 + growth_rate, months)
print(f"Projected annual revenue: ${projected:,.2f}")

# datetime: date arithmetic is surprisingly tricky without this
campaign_start = datetime.date(2025, 1, 15)
campaign_end = datetime.date(2025, 3, 31)
duration = (campaign_end - campaign_start).days
print(f"Campaign duration: {duration} days")

# random: sampling and simulation
customer_ids = [f"C{i:04d}" for i in range(1, 501)]
survey_sample = random.sample(customer_ids, 50)
print(f"\nSurvey sample (first 5): {survey_sample[:5]}")
print()

# Third-party: numpy — the backbone of numerical computing in Python
weekly_sales = np.array([12500, 11800, 13200, 15600, 14100, 13750, 16200])

print(f"Weekly sales stats:")
print(f"  Mean:   ${np.mean(weekly_sales):,.0f}")
print(f"  Std:    ${np.std(weekly_sales):,.0f}")
print(f"  Max:    ${np.max(weekly_sales):,.0f}  (day {np.argmax(weekly_sales) + 1})")
print(f"  Min:    ${np.min(weekly_sales):,.0f}  (day {np.argmin(weekly_sales) + 1})")

# 10% above average — vectorized operations across the whole array at once
above_avg = weekly_sales[weekly_sales > np.mean(weekly_sales)]
print(f"  Days above average: {len(above_avg)} of {len(weekly_sales)}")
```
