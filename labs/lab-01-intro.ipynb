### 5. Comparison Operators

We often need to compare values in our code. Python provides several comparison operators that return `True` or `False` based on the comparison.

| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `10 > 7` | `True` |
| `<` | Less than | `4 < 2` | `False` |
| `>=` | Greater than or equal to | `3 >= 3` | `True` |
| `<=` | Less than or equal to | `8 <= 6` | `False` |

````python
a = 10
b = 5
print(a > b)
print(a == b)
print("hello" != "world")
````

## Troubleshooting

**`TypeError: can only concatenate str (not "int") to str`**  
Mixing strings and numbers in concatenation. Fix: wrap the number with `str()` → `"value: " + str(42)`

**`NameError: name 'x' is not defined`**  
Using a variable before assigning it, or a typo in the name. Check spelling and make sure the assignment cell ran.

**`SyntaxError` near `=`**  
Using `=` where `==` is needed in a comparison. Remember: `=` assigns a value, `==` compares two values.

**`TypeError: unsupported operand type(s) for *: 'str' and 'int'`**  
Trying to do math on a string variable. Check `type()` of your variable and convert with `int()` or `float()` first.

**General debugging tip:** When you get an error, read the last line of the traceback first — it tells you what went wrong. Then look at the line number to find where.
