### A1: Jupyter Notebooks as Professional Tools (~8 min)

A Jupyter notebook isn't just a code runner — it's a **communication tool**. Analysts use notebooks to document methodology, present findings to stakeholders, and create reproducible workflows that others can inspect and trust.

The key to professional notebooks is the **three-cell pattern** — use it for every analysis step:

- **Markdown before the code** — explain the *why*: what question are you answering, what data are you working with, and what approach will you take?
- **Code cell** — compute the result with well-named variables and formatted output
- **Markdown after the code** — interpret the result in business terms: what does the number mean, and what action does it inform?

> **Problem → Approach → Code → Interpretation**

In Tuesday's lecture you saw two example notebooks that demonstrate this pattern:
- **[Example Analysis Notebook](https://github.com/bradleyboehmke/uc-bana-7025/blob/main/notebooks/examples/example_notebook.ipynb)** — a full business analysis with context, methodology, and findings

❌ **Undocumented** (correct answer, no story):

✅ **Documented** — three-cell pattern:

*[Markdown before]* **Q4 Net Revenue — Wireless Headphones.** We have 1,240 units ordered at $79.99 list price. Finance applies a 4% return allowance before revenue is recognized. Goal: confirm we met the $95K quarterly target.

### A2: Basic Data Structures in Python (~12 min)

Choosing the right data structure is a foundational skill. Each structure has a specific purpose:

| Structure | Ordered | Mutable | Best for |
|-----------|---------|---------|----------|
| `list` | ✓ | ✓ | Sequences that may grow, shrink, or change over time |
| `tuple` | ✓ | ✗ | Fixed groupings that should never be modified |
| `dict` | ✓* | ✓ | Labeled data where you access values by name, not position |

*Insertion order preserved in Python 3.7+

The demos below walk through each structure in a retail context. For each one, pay attention to: (1) when you would choose it, and (2) what operations it supports.

**Lists** — ordered, mutable sequences. Use when values may change over time: daily figures that need corrections, incoming records, or results you're building incrementally.

````python
# LIST: ordered and mutable — append new data, update corrections, then summarize
weekly_sales = [4250.50, 3890.25, 5120.75, 4600.00, 6340.80]

weekly_sales.append(5890.00)  # Friday's figures arrived late — add to end
weekly_sales[0] = 4275.00     # Monday's figure was corrected — update in place

print("Weekly Sales")
print(f"  Days recorded: {len(weekly_sales)}")
print(f"  Total:   ${sum(weekly_sales):,.2f}")
print(f"  Best:    ${max(weekly_sales):,.2f}")
print(f"  Worst:   ${min(weekly_sales):,.2f}")
````

**Tuples** — ordered, immutable groupings. Use for facts that are fixed by nature: GPS coordinates, record identifiers, or constant configuration values. Immutability is a feature — it signals to future readers (and to Python) that this value should never change.

````python
# TUPLE: ordered and immutable — coordinates are a geographic fact, not a variable
store_location = (39.1031, -84.5120)   # (latitude, longitude) for Cincinnati HQ

lat, lon = store_location   # tuple unpacking — assign each element to its own name
print(f"Store HQ: {lat}°N, {lon}°W")

# Tuples enforce immutability — this line would raise a TypeError:
# store_location[0] = 41.0   # ← uncomment to see the error
````

**Dictionaries** — mutable key-value pairs. Use when data has meaningful labels and you need to access or update individual fields by name, not by position. Any "row" of named attributes — a product record, a customer profile, a store summary — maps naturally to a dictionary.

````python
# DICT: labeled key-value pairs — access and update by name, not by index
product = {
    "name": "Wireless Headphones",
    "sku": "WH-2024",
    "price": 79.99,
    "in_stock": 142
}

print(f"Product: {product['name']}  |  SKU: {product['sku']}")
print(f"Price: ${product['price']}  |  Stock: {product['in_stock']} units")

product["price"] = 69.99        # update an existing field in place
product["on_clearance"] = True  # add a new field — dicts are flexible

print(f"\nUpdated: ${product['price']}  |  Clearance: {product['on_clearance']}")
````

````python
# DICT: nested structures — a product catalog with multiple entries and mixed data types
catalog = {
    "WH-2024": {
        "name": "Wireless Headphones",
        "price": 79.99,
        "in_stock": 142,
        "tags": ["electronics", "audio", "wireless"],      # list nested in dict
        "dimensions": (7.5, 6.2, 3.1)                      # tuple nested in dict
    },
    "BT-2024": {
        "name": "Bluetooth Speaker",
        "price": 49.99,
        "in_stock": 87,
        "tags": ["electronics", "audio", "portable"],
        "dimensions": (4.0, 4.0, 3.5)
    },
    "SW-2024": {
        "name": "Smart Watch",
        "price": 199.99,
        "in_stock": 34,
        "tags": ["electronics", "wearables"],
        "dimensions": (1.7, 1.5, 0.4)
    }
}

# Access a top-level value
print(f"Number of products in catalog: {len(catalog)}")

# Access a nested value — two bracket lookups: catalog[sku][field]
print(f"\nSmart Watch price: ${catalog['SW-2024']['price']}")

# Access a value nested inside a list inside a dict
print(f"First tag for BT Speaker: {catalog['BT-2024']['tags'][0]}")

# Access a value nested inside a tuple inside a dict
height = catalog['WH-2024']['dimensions'][2]
print(f"Headphone height: {height} in")
````

### A3: Importing and Using Libraries (~10 min)

Python's library ecosystem is organized in two tiers:

- **Standard library** — ships with every Python installation; no install required (`datetime`, `random`, `math`, `os`)
- **Third-party libraries** — installed separately via `pip`; power the data science ecosystem (`numpy`, `pandas`, `seaborn`, `scikit-learn`)

The demos below work through both tiers. The standard library handles everyday tasks out of the box; third-party packages extend Python into specialized domains that would be impractical to build from scratch.

**Standard Library: `datetime`** — date arithmetic for business calendars, deadlines, and time-based reporting. No install needed — it ships with Python.

````python
import datetime

today = datetime.date.today()
fiscal_year_end = datetime.date(today.year, 12, 31)
days_remaining = (fiscal_year_end - today).days

print(f"Today:                         {today}")
print(f"Fiscal year end:               {fiscal_year_end}")
print(f"Days remaining in fiscal year: {days_remaining}")
````

**Standard Library: `random`** — generate random values for simulations, sampling, and scenario modeling. Also ships with Python.

````python
import random

# Simulate promotional discount offers for three loyalty program customers
offer_a = random.randint(5, 25)
offer_b = random.randint(5, 25)
offer_c = random.randint(5, 25)

offers = [offer_a, offer_b, offer_c]
print(f"Discount offers generated: {offers}%")
print(f"Average discount offered:  {sum(offers) / len(offers):.1f}%")
````

**NumPy** — the backbone of numerical computing in Python. Its defining feature is **vectorized math**: operations apply to every element of an array simultaneously, without a loop.

````python
import numpy as np

sales_array = np.array([4250.50, 3890.25, 5120.75, 4600.00, 6340.80])

projected = sales_array * 1.10   # 10% growth applied to all elements at once

print(f"Actual sales:     {sales_array}")
print(f"Projected (+10%): {projected.round(2)}")
print(f"Mean:  ${np.mean(sales_array):,.2f}")
print(f"Stdev: ${np.std(sales_array):,.2f}")
print(f"Total: ${np.sum(sales_array):,.2f}")
````

## Troubleshooting

**`IndexError: list index out of range`** — You accessed an index that doesn't exist. Python uses zero-based indexing: a 5-item list has valid indices 0–4, not 1–5.

**`KeyError`** — The dictionary key you tried doesn't exist. Check available keys with `dict.keys()`, or use `dict.get('key', default)` for safe access without an error.

**`TypeError: 'tuple' object does not support item assignment`** — Tuples are immutable. To modify grouped data, use a list or create a new tuple.

**`AttributeError: 'list' object has no attribute '...'`** — You called a NumPy method on a plain Python list. Convert first: `np.array(my_list)`.

**`ModuleNotFoundError`** — The package isn't installed in this environment. Run `!pip install package_name` in a code cell, then re-run the import.
